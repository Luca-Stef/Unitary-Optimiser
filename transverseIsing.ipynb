```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
def init_ising_H_bond(g,h,J):
    """ Returns bond hamiltonian"""
    sx = np.array([[0.,1.],[1.,0.]])
    sz = np.array([[1.,0.],[0.,-1.]])
    d = 2
    H = -J*np.kron(sz,sz)
    H = H + g/2.*(np.kron(sx,np.eye(d))+np.kron(np.eye(d),sx))
    H = H + g/2.*(np.kron(sx,np.eye(d))+np.kron(np.eye(d),sx))
    H = H + h/2.*(np.kron(sz,np.eye(d))+np.kron(np.eye(d),sz))
    return np.reshape(H,(d,d,d,d))
```

```python
J = 1
g = 0
Points = 10
Eexact = []
T = []
r = 0
while r < Points:
    g += 0.5 #2/Points
    #H = init_ising_H_bond(g/2,0,J)
    f = lambda k,g : -2*np.sqrt(1+g**2-2*g*np.cos(k))/np.pi/2.
    E0_exact = integrate.quad(f, 0, np.pi, args=(g,))[0]
    Eexact.append(E0_exact)
    T.append(g)
    r += 1

plt.plot(T,Eexact)
plt.show()
```
